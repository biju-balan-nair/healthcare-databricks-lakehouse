```python
import pandas as pd
import numpy as np
from faker import Faker
import random
import os

fake = Faker()

os.makedirs("../datasets/raw/emr", exist_ok=True)

patients = []

for i in range(1000):

    patients.append({
        "patient_id": f"P{i+1:05}",
        "name": fake.name(),
        "gender": random.choice(["Male", "Female"]),
        "dob": fake.date_of_birth(minimum_age=18, maximum_age=90),
        "city": fake.city()
    })

patients_df = pd.DataFrame(patients)

patients_df.to_csv(
    "../datasets/raw/emr/patients.csv",
    index=False
)

print("patients.csv generated successfully")
```
